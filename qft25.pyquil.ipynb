```python
from pyquil import Program, get_qc
from pyquil.gates import CPHASE, H, SWAP, MEASURE
import numpy as np
import time
```

```python
p = Program()
```

```python
ro = p.declare('ro', memory_type='BIT', memory_size=25)
```

```python
nbr_qbs=24

#buld the circuit
for i in range(nbr_qbs,0,-1): #print(i)
    p.inst(H(i))
    for j in range(nbr_qbs,i-1,-1):
        k=i-1; m=2**(j-k); #print(m,k,j)
        p.inst(CPHASE(np.pi / m, k, j))
        
p.inst(H(0));#print(i-1)

#swap
m=int((nbr_qbs+1)/2)
for i in range(0,m):
    j=nbr_qbs-i; #print(i,j)
    p.inst(SWAP(i, j))

#measure
for i in range(0,nbr_qbs+1): #print(i)
    p.inst(MEASURE(i, ro[i]))    
```
