```python
from qiskit import QuantumRegister, ClassicalRegister
from qiskit import QuantumCircuit, execute, Aer

import numpy as np
import time

# import basic plot tools
from qiskit.tools.visualization import plot_histogram, circuit_drawer
```

```python
qc = QuantumCircuit()
```

```python
nbr_qbs=24
q = QuantumRegister(nbr_qbs+1, 'q')
c = ClassicalRegister(nbr_qbs+1, 'c')
```

```python
qc.add_register(q)
qc.add_register(c)
```

```python
#buld the circuit
for i in range(nbr_qbs,0,-1): #print(i)
    qc.h(q[i])
    for j in range(nbr_qbs,i-1,-1):
        k=i-1; m=2**(j-k); #print(m,k,j)
        qc.cu1(np.pi / m, q[k], q[j])
        
qc.h(q[0]);#print(i-1)

#swap
m=int((nbr_qbs+1)/2)
for i in range(0,m):
    j=nbr_qbs-i; #print(i,j)
    qc.swap(q[i], q[j])
    
#measure
for i in range(0,nbr_qbs+1): #print(i)
    qc.measure(q[i], c[i])    
    
 
```
